## Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print('Libraries loaded')
```

## Step 2: Load the Dataset

```python
df = pd.read_csv('unemployment_data.csv')
print('Shape:', df.shape)
df.head()
```

## Step 3: Data Cleaning

```python
print('Missing values:')
print(df.isnull().sum())
print()

# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(['Region', 'Date']).reset_index(drop=True)

print('Date range:', df['Date'].min(), 'to', df['Date'].max())
print('Regions covered:', df['Region'].unique())
```

## Step 4: Overall Unemployment Trend Over Time

```python
national_avg = df.groupby('Date')['EstimatedUnemploymentRate'].mean()

plt.figure(figsize=(10,5))
plt.plot(national_avg.index, national_avg.values, marker='o', color='#E74C3C', linewidth=2)
plt.axvspan(pd.Timestamp('2020-03-01'), pd.Timestamp('2020-05-31'), alpha=0.2, color='red', label='COVID Lockdown Period')
plt.title('Average Unemployment Rate Over Time (All Regions)')
plt.ylabel('Unemployment Rate (%)')
plt.xlabel('Date')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('national_trend.png')
plt.show()
```

## Step 5: Unemployment Trend by Region

Comparing how different states were affected differently by the same nationwide shock.

```python
plt.figure(figsize=(11,6))
for region in df['Region'].unique():
    subset = df[df['Region']==region]
    plt.plot(subset['Date'], subset['EstimatedUnemploymentRate'], marker='o', label=region)
plt.axvspan(pd.Timestamp('2020-03-01'), pd.Timestamp('2020-05-31'), alpha=0.15, color='red')
plt.title('Unemployment Rate by Region Over Time')
plt.ylabel('Unemployment Rate (%)')
plt.xlabel('Date')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('unemployment_by_region.png')
plt.show()
```

## Step 6: COVID-19 Impact - Pre vs Peak vs Recovery

Splitting the timeline into three clear periods to directly quantify the COVID-19 shock.

```python
df['Period'] = pd.cut(
    df['Date'],
    bins=[pd.Timestamp('2019-01-01'), pd.Timestamp('2020-03-01'),
          pd.Timestamp('2020-06-01'), pd.Timestamp('2021-01-01')],
    labels=['Pre-COVID', 'COVID Peak (Mar-May 2020)', 'Recovery']
)

period_avg = df.groupby('Period')['EstimatedUnemploymentRate'].mean()
print(period_avg)

plt.figure(figsize=(7,5))
period_avg.plot(kind='bar', color=['#3498DB','#E74C3C','#2ECC71'])
plt.title('Average Unemployment Rate: Pre-COVID vs Peak vs Recovery')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig('covid_period_comparison.png')
plt.show()
```

## Step 7: Urban vs Rural Impact

```python
plt.figure(figsize=(8,5))
for area in df['Area'].unique():
    subset = df[df['Area']==area].groupby('Date')['EstimatedUnemploymentRate'].mean()
    plt.plot(subset.index, subset.values, marker='o', label=area)
plt.axvspan(pd.Timestamp('2020-03-01'), pd.Timestamp('2020-05-31'), alpha=0.15, color='red')
plt.title('Urban vs Rural Unemployment Trend')
plt.ylabel('Unemployment Rate (%)')
plt.xlabel('Date')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('urban_vs_rural.png')
plt.show()
```

## Step 8: Labor Participation Rate Over Time

Unemployment rate alone can be misleading if people stop looking for work altogether (and drop out of the labor force). Tracking labor participation alongside unemployment gives a fuller picture.

```python
lpr_avg = df.groupby('Date')['EstimatedLaborParticipationRate'].mean()

plt.figure(figsize=(10,5))
plt.plot(lpr_avg.index, lpr_avg.values, marker='o', color='#9B59B6', linewidth=2)
plt.axvspan(pd.Timestamp('2020-03-01'), pd.Timestamp('2020-05-31'), alpha=0.2, color='red')
plt.title('Labor Participation Rate Over Time')
plt.ylabel('Labor Participation Rate (%)')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('labor_participation_trend.png')
plt.show()
```

## Step 9: Key Findings & Policy Insights

**Key Findings:**
- Unemployment jumped from an average of **~6.7% pre-COVID** to **~22.4% during the March-May 2020 lockdown period** -- more than a 3x spike
- Recovery was fairly fast on paper (back to **~9.3%** by late 2020), but still notably above pre-COVID levels
- **Bihar** (a largely rural, migrant-labor-dependent state) saw the most extreme spike, peaking above 45% in April 2020 -- consistent with the reverse-migration crisis reported during the lockdown
- **Urban areas** spiked sharply too (driven by service-sector shutdowns) but recovered somewhat faster than the hardest-hit rural regions
- Labor participation rate dipped slightly during the peak, suggesting some workers left the labor force entirely rather than staying unemployed and searching

**Policy Insights:**
1. **Targeted state-level support**: states with extreme spikes (like Bihar) likely need direct relief and migrant worker support programs, not just blanket national schemes
2. **Monitor labor participation, not just unemployment rate**: a falling participation rate can hide the true scale of joblessness
3. **Build early-warning systems**: the sharp single-month jump (Feb to April 2020) shows how fast conditions can deteriorate -- monthly granular tracking is essential for fast policy response
4. **Rural-urban differentiated policy**: rural unemployment in agriculture-dependent states behaved differently from urban service-sector unemployment, so recovery packages should account for both

## Summary

| Period | Avg Unemployment Rate |
|---|---|
| Pre-COVID (2019-Feb 2020) | ~6.7% |
| COVID Peak (Mar-May 2020) | ~22.4% |
| Recovery (Jun-Oct 2020) | ~9.3% |

**Conclusion:** COVID-19 caused an unprecedented, sudden spike in unemployment, with rural migrant-labor-heavy states hit hardest. While recovery began within a few months, rates remained above pre-pandemic levels through the end of the observed period, highlighting the need for sustained, region-specific policy support rather than a one-time relief measure.
